```python
import json
import pandas as pd
```

```python
f = open('product_cats.json')
data = json.load(f)
data
```

```python
product_cats = []
for cat in data:
    product_cats.append(cat['displayName'])
product_cats
```

```python
df = pd.DataFrame({'Product Categories': product_cats})
df
```

```python
df.to_csv('./produc_categories.csv', index = False)
```
