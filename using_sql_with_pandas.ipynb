```python
import pandas as pd
from pandasql import sqldf
mysql = lambda q: sqldf(q, globals())
```

```python
df = pd.read_csv('./data/penguins.csv')
mysql("SELECT * from df")
```

```python
adelie_penguins = mysql("SELECT * from df where species = 'Adelie Penguin (Pygoscelis adeliae)'")
adelie_penguins
```

```python
adelie_penguins.to_csv('adelie_penguins.csv')
```
